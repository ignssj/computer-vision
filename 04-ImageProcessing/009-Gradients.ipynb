```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('/home/wanakin/DATA/sudoku.jpg',0)
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
display_img(sobely)
```

```python
blended = cv2.addWeighted(src1=sobelx,alpha=0.5,src2=sobely,beta=0.5,gamma=1)
```

```python
display_img(blended)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
display_img(laplacian)
```

```python
ret,th1 = cv2.threshold(blended,127,255,cv2.THRESH_BINARY)
display_img(th1)
th1.shape
```

```python
kernel = np.ones((3,3),dtype=np.uint8)
opening = cv2.morphologyEx(th1,cv2.MORPH_OPEN,kernel,iterations=1)
```

```python
display_img(opening)
```

```python
gradient = cv2.morphologyEx(opening,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient)
```
